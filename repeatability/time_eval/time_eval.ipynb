```python
import pandas as pd
```

```python
df = pd.read_csv("./CT2_times.csv",sep=",")
df
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
df.mean()
```

```python
# Draw a nested violinplot and split the violins for easier comparison
ax = sns.violinplot(data=df, linewidth=1, cut=1, inner='box')
ax.set(
    xlabel="{:.2f} repetitions".format(df["Times"].count()),
    ylabel="time [s]",
    title="Time distribution, mean: {:.2f}".format(df["Times"].mean())
)
sns.despine(left=True)
```

```python
ax.figure.savefig("time_dist_CT2.png")
```
