```python
import pandas as pd
```

```python
df = pd.read_csv("./CT4_score_parameter_evalTH_5.csv",sep=",")
df['th_intDen']=(df['th_intDen']/df['th_intDen'].max())*1
df
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
#cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
cmap = sns.color_palette("coolwarm", as_cmap=True)
g = sns.relplot(
    data=df,
    x="thL", y="thU",
    hue="Score", size="th_intDen",
    palette=cmap, sizes=(10, 200),
)
g.set(
    xlabel="Lower threshold",
    ylabel="Upper threshold",
    title="Score evaluation over whole brightness settings"
)
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

```python
g.figure.savefig("CT4_int_score_th_range.png")
```
